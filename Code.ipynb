```python
#각 제품별 주차별 요일 주문량 피벗 테이블 생성
table=pd.pivot_table(data,index=['ITEM_CD','week','day'], values=['ITEM_QTY'],aggfunc='sum')
table
```

```python
#모델 학습에 필요한 데이터셋 생성
item_list=data['ITEM_CD'].unique()
from tqdm import tqdm 
import time

tmp_list=[]
for i in tqdm(item_list):
    for j in range(1,18):
        tmp={}
        try:
            temp=table.loc[i].loc[j]
            if len(temp)==3:
                tmp['target']=temp.loc[0][0]
                tmp['SAT']=temp.loc[5][0]
                tmp['SUN']=temp.loc[6][0]
                tmp['ITEM_CD']=i                  
            else:
                continue
  
            tmp_list.append(tmp)
        except:
            continue
```

```python
df_final=pd.DataFrame(tmp_list)
df_final
```

```python
CORP_ID={}
BKG_TYP={}
PRICE={}
DLVCLSFCD={}
POST_ZONE={}
from scipy import stats
for item in tqdm(df_final['ITEM_CD'].unique()):
    CORP_ID[item]=data[data['ITEM_CD']==item]['CORP_ID'].iloc[0]
    BKG_TYP[item]=stats.mode(data[data['ITEM_CD']==item]['BKG_TYP'])[0][0]
    PRICE[item]=data[data['ITEM_CD']==item]['ITEM_AMT'].iloc[0]/data[data['ITEM_CD']==item]['ITEM_QTY'].iloc[0]
    DLVCLSFCD[item]=stats.mode(data[data['ITEM_CD']==item]['DLVCLSFCD'])[0][0]
    POST_ZONE[item]=stats.mode(data[data['ITEM_CD']==item]['POST_ZONE'])[0][0]
```

```python
for i in tqdm(range(len(df_final))):
    df_final.loc[i,'CORP_ID']=CORP_ID[df_final.loc[i,'ITEM_CD']]
    df_final.loc[i,'BKG_TYP']=BKG_TYP[df_final.loc[i,'ITEM_CD']]
    df_final.loc[i,'price']=PRICE[df_final.loc[i,'ITEM_CD']]
    df_final.loc[i,'DLVCLSFCD']=DLVCLSFCD[df_final.loc[i,'ITEM_CD']]
    df_final.loc[i,'POST_ZONE']=POST_ZONE[df_final.loc[i,'ITEM_CD']]
```

```python
# 범주형 변수 type변환
data['ITEM_CD'] = data['ITEM_CD'].astype("category")
data['BKG_TYP'] = data['BKG_TYP'].astype("category")
data['CORP_ID'] = data['CORP_ID'].astype("category")
data['DLVCLSFCD'] = data['DLVCLSFCD'].astype('category')
data['POST_ZONE'] = data['POST_ZONE'].astype('category')

```

```python
# 인코딩
data['DLVCLSFCD'] = data['DLVCLSFCD'].cat.codes
data['POST_ZONE'] = data['POST_ZONE'].cat.codes
data['ITEM_CD'] = data['ITEM_CD'].cat.codes
data['BKG_TYP'] = data['BKG_TYP'].cat.codes
data['CORP_ID'] = data['CORP_ID'].cat.codes
```

```python
# 변환 확인
data.info()
```

```python
def outliers_iqr(data):
    q1, q3 = np.percentile(data['target'], [25,75])
    iqr=q3-q1
    lower_bound=q1-(iqr*1.5)
    upper_bound=q3+(iqr*1.5)
    return data[(data['target']>lower_bound)&(data['target']<upper_bound)]
```

```python
# 아웃라이어 제거
data=outliers_iqr(data)
```

```python
data =data.reset_index()
data
```

```python
# 종속, 독립 변수 분리
y = data['target']
X = data[['SAT','SUN','price','ITEM_CD','CORP_ID','BKG_TYP','DLVCLSFCD','POST_ZONE']]
```

```python
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size=0.2, random_state=1422)
print(X_train.shape, X_test.shape)
```

```python
# catboost 설치 및 fitting
!pip install catboost
from catboost import CatBoostRegressor

model_cat = CatBoostRegressor(iterations=2000, l2_leaf_reg = 0.003, random_seed = 1422)
model_cat.fit(X_train,y_train)
```

```python
pred_cat = model_cat.predict(X_test)
pred_cat
```

```python
print('The rmse of prediction is:', round(mean_squared_error(pred_cat, y_test) ** 0.5, 5))
```
